```python
C = [[2, 1], [1, 2]] #define the matrix
print(C)             #print matrix C
A = np.linalg.cholesky(C) #perform Cholesky decomposition on C. (Find A such that, C = A @ A.T)
print(A)             #print matrix A
print(A @ A.T)       #this will be equal to C
```

```python
theta = np.pi / 3 #define angle theta in radians
u = [np.sin(theta), np.cos(theta)] #create a unit vector
print("The vector: ", u)           #print the unit vector
print("Sum of squares: ", u[0]**2 + u[1]**2) #calculate sum of squares. This will be 1 in unit vectors
print("Degrees: ", theta*180 / np.pi) #convert angle in radians to degrees
```

```python
nPoints = 50;                              #initialize the number of points
pVars = np.zeros(nPoints)                  #array to store the variance of projections
thRange = np.linspace(0, 2*np.pi, nPoints) #create an array of nPoints equally spaced angles between 0 and 2*pi

#calculating the variance of projections
for n in range(nPoints):
    theta = thRange[n]
    u = [np.sin(theta), np.cos(theta)]
    pVars[n] = np.var( Y @ u )

#plot the graph
plt.plot(pVars)
plt.grid(True)
plt.xlabel("Direction", fontsize=14)
plt.ylabel("Variance of Projections", fontsize=14)
```
